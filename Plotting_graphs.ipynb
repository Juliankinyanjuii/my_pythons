#plotting line graph


```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import plotly.express as px
```

```python
x=[10,13,16,17,17,18,20,24,26,25,25,27,17,12,13,28,29,34]
x.sort()
y=[11,12,15,16,16,19,18,23,25,24,24,28,28,11,12,27,29,33]
y.sort()
```

```python
#plotting a line graph using matplotlib
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel('y-axis')
plt.title("line graph")
plt.show()
```

```python
#plotting line graph using seaborn
sns.lineplot(x=x,y=y, color="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" a line graph")
plt.show()
```

#pie charts

```python
animals=["Dogs","elephants","cats","donkeys", "chicken"]
counts=[10,20,30,40,50]
```

```python
#plotting a pie chart using matplotlib
plt.pie(counts,labels=animals, autopct='%1.3f%%',startangle=90)
plt.show()
```

#histogram

```python
#using matplotlib
plt.hist(y,bins=5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Histogram")
plt.show()
```

```python
#using seaborn
sns.histplot(y,bins=5, color="green")
plt.xlabel("x axis")
plt.ylabel("y axis ")
plt.title("Histogram using seaborn")
plt.show()
```

#Scatterplot

```python
#using matplotlib
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" a scatterplot using matplotlib")
plt.show()
```

```python
sns.scatterplot(x=x,y=y,color="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" A scatter plot using seaborn")
plt.show()
```

#Bargraph

```python
#using matplotlib
plt.bar(x,y, color="pink")
plt.title("this is a bargraph using matplotlib")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# drawing a bargraph using seaborn
sns.barplot(x=x,y=y, color="brown")
plt.title("a bar graph  using seaborn")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
#drawing a countplot using seaborn
sns.countplot(y=counts,hue=animals)
plt.xlabel(" x axis")
plt.ylabel("y axis")
plt.title(" a countplot")
plt.show()
```
